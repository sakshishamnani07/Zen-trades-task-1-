```python
import requests
import pandas as pd

# Fetch data from the API
url = 'https://s3.amazonaws.com/open-to-cors/assignment.json'
response = requests.get(url)
data = response.json()

# Print the entire data to understand its structure
# Assuming 'products' is the key containing the list of products
products_data = data.get('products', [])
sorted_data = sorted(products_data.values(), key=lambda x: int(x.get('popularity', 0)), reverse=True)

    # Display data in a Pandas DataFrame
df = pd.DataFrame(sorted_data, columns=['title', 'price', 'popularity', 'subcategory'])
df
```
